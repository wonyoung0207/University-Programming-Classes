  3. **rep(repeat, 반복 벡터), seq(sequence, 순열 벡터)**


* rep

````
## "a"을 10번 반복
rep("a", times = 10)
 
## "1"을 15번 반복
rep(1, times = 15)

# "a" "1"을 5번 반복
rep( c("a", 1), 5)

## "a"를 먼저 5번 반복하고, "1"을 10번 반복
rep( c("a", 1), c(5,10))
````

* seq

````
# 2부터 5까지
seq(2, 5) # 2:5

# 3부터 10까지
seq(from = 3, to = 10) 

# 2부터 10까지 2씩 커짐
seq(from = 2, to = 10, by = 2)

# 2부터 10까지, 수열의 길이는 3
seq(from = 2, to = 10, length.out = 3)
````

## **(4) 배열/데이터프레임/리스트**

* 배열(array): 열과 행으로 구성된 다차원의 데이터(단일 데이터형), 벡터의 모임


### **1. 배열**

* 배열 생성 함수 ( array(), matrix() )

```
배열 Array 생성하기 array() 같은타입의 벡터 모임

array(data = NA, dim= length(data), dimnames = NULL)

dim : 차원 생성

dimnames : 차원별 이름
```

````
# dim = c(행의 길이, 열의 길이, 깊이, ...)
ar1 <- array(1:3, dim = c(5)) # 1차원 == 벡터
ar2 <- array(1:3, dim = c(2,4)) # 2차원
ar3 <- array(1:3, dim = c(2,4,4)) # 3차원
# array(1:3, c(2,4,3,5)) # 4차원 배열   
````

````
print(ar1) # 1차원 == 벡터
````

````
print(ar2) # 2차원
````

````
print(ar3) # 3차원
````

* indexing

````
ar2[1, 3] # 1 행, 3 열의 값
ar2[2, 2] # 2 행, 2 열의 값
````

````
ar3[1, 3, 1] # 1 번째 깊이의 1 행 3 열 값
ar3[1, 2, 4] # 4 번째 깊이의 1 행 2 열 값
````

```
행렬 Matrix 생성하기 matrix() 2차원배열

matrix(data = NA, nrow = 1, ncol =1, byrow = F, dimnames = NULL)

nrow : 행 요소의 개수

ncol : 열 요소의 개수

byrow : 데이터를 행 단위로 배치할 지 여부

dimnames : 행과 열의 이름 list
```

````
temp<- 1:12
print(temp)
````

````
m <- matrix(temp,nrow = 3) # or ncol = 4

print(m)
````

````
# row부터 채워나간다 기본은 F col부터 채워나감
m1 <- matrix(temp,nrow =3, byrow=T)

print(m1)
````

* 행렬 합치기 cbind(열), rbind(행)

````
v1 <- 1:4
v2 <- 5:8
v3 <- 9:12
````

````
# 열 합치기
print(cbind(v1,v2,v3))
````

````
# 행 합치기
print(rbind(v1,v2,v3))
````

* 배열 연산

자세한 내용 : https://thebook.io/006723/ch02/06/03/

  1. apply

    (1) apply 상세설명 : https://issactoast.com/53
    
```
  apply(X, MARGIN, FUN, ...)

X : 적용할 배열 또는 행렬

MARGIN : 함수를 적용할 차원, 행렬의 경우 1이면 행별, 2이면 열별로 적용

FUN : 적용할 함수
```


````
v1 <- 1:4
v2 <- 5:8
v3 <- 9:12

m2 <- cbind(v1,v2,v3)
print(m2)
````

````
# 각 행별로 sum(합)
print(apply(m2,1,sum))

# 각 열별로 sum(합)
print(apply(m2,2,sum))
````

````
# 각 행별로 mean(평균)
print(apply(m2,1,mean))

# 각 열별로 mean(평균)
print(apply(m2,2,mean))
````
